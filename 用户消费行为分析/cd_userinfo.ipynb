```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from matplotlib_inline import backend_inline
%matplotlib inline
backend_inline.set_matplotlib_formats('svg')
plt.style.use('ggplot') # 更改风格，选用R语言绘图库的风格
# matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号
```

```python
# 导入数据
columns = ['user_id', 'order_dt', 'order_products', 'order_amount']
df = pd.read_table('./CDNOW_master.txt', names = columns, sep = '\s+') # 分隔符是\s空格，\s+表示任意空格
df.head()

# 日期格式需要转换
# 存在一个用户一天内多次购买行为
```

```python
df.describe()
```

> (1). 用户平均每笔订单购买2.4个商品，标准差2.3，稍微有些波动，属于正常<br/>
> (2). 75%分位数说明绝大多数订单的购买量都不多，围绕在2-3个产品左右<br/>
> (3). 购买金额：反映出绝倒部分订单消费金额集中在中小额度，30-40左右

```python
df.info()
```

```python
# 对order_dt进行类型转换，转换为日期格式
# format参数：按照指定的格式去匹配要转换的数据列
df['order_date'] = pd.to_datetime(df['order_dt'], format = '%Y%m%d')
# 月份列
df['month'] = pd.to_datetime(df['order_date']).dt.to_period('M')
df.head()
```

# 用户整体消费趋势分析（按月份）

```python


plt.figure(figsize = (15, 10))
# 绘制每月的产品购买数量
d = df.groupby('month')['order_products'].sum()
ax1 = plt.subplot(221)
plt.title('每月产品的购买数量')
d.plot()

# 每月的消费金额
d2 = df.groupby('month')['order_amount'].sum()
ax2 = plt.subplot(222)
plt.title('每月的消费额')
d2.plot()

# 每月的消费次数
d3 = df.groupby('month')['user_id'].count()
ax3 = plt.subplot(223)
plt.title('每月的消费次数')
d3.plot()

# 每月的消费人数(user_id去重统计，再计算)
d4 = df.groupby('month')['user_id'].apply(lambda x : len(x.drop_duplicates()))
ax4 = plt.subplot(224)
plt.title('每月消费人数')
d4.plot()

plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.show()
```

> 前三个月的销量非常高，之后较为稳定，并且呈现下降趋势<br/>
> 前三个月消费金额较高，之后下降严重<br/>
> 在我国，1，2，3月份处于春节前后，公司可能在这段时间加大了营销力度<br/>
> 前三个月消费次数在10000左右，而后续在两千五左右<br/>
> 前三个月消费人数较多，而后下降到2000左右，趋于平稳<br/>
> 前三个月消费人数在8000 - 10000在，后续平均消费人数在2000人左右<br/>
> **总结：所有数据显示：97年前三月消费异常，后续趋于平稳**
