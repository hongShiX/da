# 一、数据清洗

### 重复值，缺失值，格式调整


#### 订单号，存在缺失值，但是量不大，可以删除

#### 订单行，对分析无关紧要，可以删除
#### 销售金额格式不对，需要转换为float

```python
# 删除重复记录
df.drop_duplicates(keep = 'first', inplace = True)

# 删除缺失值 (删除带有na的一整行数据)
df.dropna(axis = 0, how = 'any', inplace = True)

# 删除[订单行]这一列
df.drop(columns = ['订单行'], inplace = True, axis = 1)

# 重置索引(把原来的索引index删除，重置一下索引)
df.reset_index(drop = True, inplace = True)

print(df.info())
```

```python
# 修改金额格式
# 编写自定义过滤函数
def change_money(num):
    num = num.replace(',', '')
    if num.find('万元') != -1: # 找到带有万元的数据
        num = float(num[:-2]) * 10000
    elif num.find('元') != -1:
        num = float(num[:-1])
    return num
```

```python
# 执行自定义函数
df['销售金额'] = df['销售金额'].apply(change_money)
```

```python
df.head()
```

#### 异常值处理

```python
df.describe()
```

```python
# 少量销售金额为0的，删除
df = df[df['销售金额'] != 0]
df.describe()
```

```python
# 销售金额和数量存在严重的右偏现象
 
```

# 二、数据规整

新增一个辅助列：月份

```python
df['销售时间'] = pd.to_datetime(df['销售时间'])
df['销售月份'] = df['销售时间'].apply(lambda x : x.month)
df
```

# 三、 数据分析以及可视化

## 1. 配送服务是否存在问题

### a.月份维度

```python
# 对货品交货状况这一列去空格
df['货品交货状况'] = df['货品交货状况'].str.strip()
```

```python
# 对销售月份进行分组
# unstack() 以行列的形式显示
d1  = df.groupby(['销售月份', '货品交货状况']).size().unstack()

# 新增按时交货率列
d1['按时交货率'] = d1['按时交货'] / (d1['按时交货'] + d1['晚交货'])
d1

```

>从按时交货率来看，第四季度低于第三季度，可能是气候原因

### b.销售区域维度

```python
d1  = df.groupby(['销售区域', '货品交货状况']).size().unstack()
d1['按时交货率'] = d1['按时交货'] / (d1['按时交货'] + d1['晚交货'])
d1.sort_values(by = '按时交货率', ascending = False) # 按照按时交货率降序排列
```

> 西北地区存在严重的延时交货问题，需要解决

### c.货品维度

```python
d1  = df.groupby(['货品', '货品交货状况']).size().unstack()
d1['按时交货率'] = d1['按时交货'] / (d1['按时交货'] + d1['晚交货'])
d1.sort_values(by = '按时交货率', ascending = False) # 按照按时交货率降序排列
```

> 货品4存在严重的延时交货问题

### d.货品和销售区域结合

```python
d1  = df.groupby(['货品', '销售区域', '货品交货状况']).size().unstack()
d1['按时交货率'] = d1['按时交货'] / (d1['按时交货'] + d1['晚交货'])
d1.sort_values(by = '按时交货率', ascending = False) # 按照按时交货率降序排列
```

> 销售区域：最差在西北地区，货品有1和4<br/>
> 货品：最差的货品2，主要送往华东和马来西亚，主要是马来西亚的送货较晚导致

## 2. 是否存在尚有潜力的销售区域

### a.月份维度

```python
d1 = df.groupby(['销售月份', '货品'])['数量'].sum().unstack()
d1.plot(kind = 'line')
```

> * 货品2在10月份或12月份，销售猛增，原因猜测有二：
>   * 1. 公司加大了营销力度
>   * 2. 开发了新的市场（后面得出的结论是，公司没有开发新的市场，只是增大了营销力度）

### b.不同区域 

```python
d1 = df.groupby(['销售区域', '货品'])['数量'].sum().unstack()
d1
```

> 从销售区域看，每种货品销售区域为1~3个，货品1有三个销售区域，货品2有两个销售区域，其余货品均有1个销售区域
