### 处理缺失数据 

```python
# 定义一个收集NaN的个数的函数
def get_null_record(df):
    dict_null = {}
    for column in df.columns:
        if not len(df[df[column].isnull()]) == 0:
            dict_null[column] = len(df[df[column].isnull()])
    return dict_null

```

```python
data_null_dic = get_null_record(df)
print (data_null_dic)
```

#### 处理船舱的缺失值
船舱Cabin字段的NaN数量太大，将这些缺失值用'UnKnown'填充

```python
df['Cabin'].fillna('Unknown', inplace = True)
```

####  处理年龄的缺失值
用年龄的平均值来填充这些缺失值

```python
mean_age = df['Age'].mean()
mean_age = int(mean_age)

df['Age'].fillna(mean_age, inplace = True)
```

#### 处理登录港口的缺失值
用S填充

```python
df['Embarked'].fillna('S', inplace = True)
```

# 问题1：年龄与乘客生还的情况怎么样

#### 1.1生还数和未生还数 

```python
s = df.groupby('Survived')['PassengerId'].count() # 根据生还与否分组查询人数
t = ['未生还', '生还']
v = s.values
bar_colors = ['tab:red', 'tab:green']
fig, ax = plt.subplots()
plt.title('未生还数与生还数统计')
ax.set_ylabel('生还数')

ax.bar(t, v, width = 0.5, color = bar_colors)
```

从条形图中看，死亡人数高于生还人数

#### 1.2 年龄分布

```python
df['Age'].hist()

plt.title('年龄直方图', fontsize = 20)

plt.grid(False)
```

大部分乘客年龄在20-40之间

#### 1.3 年龄段与生还的相关系数

```python
# 计算最大年龄
age_max = df['Age'].max()
age_max
```

```python
# 计算最小年龄
age_min = df['Age'].min()
age_min
```

```python
# 对age分组，间隔为12， 年龄段将被分成['(0, 12]', '(12, 24]', '(24, 36]', '(36, 48]', '(48, 60]', '(60, 72]', '(72, 84]']
bins = np.arange(0, 90, 12)
bins
```

```python
# 新增一个变量Age_group,对应每个Age变量所属年龄段标签
df['Age_group'] = pd.cut(df['Age'], bins)
```

```python
df.head()
```

```python
# 计算生还与否和年龄段的相关性
r = df['Survived'].corr(df['Age_group'], method = 'kendall')
r
```

年龄段与生还的相关系数很小，相关性很弱，相关系数在此仅供参考

#### 1.4 各个年龄段和乘客生还的数据

```python
# 统计每个年龄段的总人数
df_total_count_age_group = df.groupby('Age_group', as_index = False)['PassengerId'].count()
df_total_count_age_group
```

```python
# 计算各个年龄段的生还人数
df_sur_count_age_group = df[df['Survived'] == 1].groupby(['Age_group'], as_index = False)['PassengerId'].count()
df_sur_count_age_group
```

```python
# 计算各个年龄段的生还率
suivival_rate_each_group = df_sur_count_age_group['PassengerId'] / df_total_count_age_group['PassengerId']
suivival_rate_each_group = pd.Series(suivival_rate_each_group.values, index = df_sur_count_age_group['Age_group'])
suivival_rate_each_group
```

#### 1.5 可视化年龄段生还率

```python
# 可视化每个年龄段的生还人数

# fig, ax = plt.subplots()
df.groupby(['Age_group'])['PassengerId'].count().plot(kind = 'bar')
plt.title('各个年龄段生还人数')
plt.ylabel('生还人数')
```

生还人数最多的年龄段在24-36岁之间

```python
# 可视化各个年龄段的生还率
suivival_rate_each_group.plot(kind='bar')
plt.xlabel('年龄段')
plt.ylabel('生还率')
plt.title('各个年龄段的生还率')
```

在(0, 12]的乘客的生还率最高， 在(60, 72]年龄段的乘客生还率最低

>结论：
>1. 死亡人数比生还人数多
>2. 生还与否和年龄段的相关性很小
>3. 生还人数最多的年龄段在24-36岁之间
>4. 在(0, 12]的乘客的生还率最高， 在(60, 72]年龄段的乘客生还率最低,其他年龄段的生还率相差不是很大

# 问题2：性别与乘客生还的情况怎么样？

#### 2.1 性别总计

```python
df.head()
```

```python
df_sex = df.groupby('Sex')['PassengerId'].count()
df_sex.replace({'female': '女', 'male': '男'}, inplace = True)
df_sex
```

```python
df_sex.plot(kind = 'bar', title = '男女乘客人数', xlabel = '性别', ylabel = '人数')
```

从条形图可以看出，男性乘客要多于女性乘客

#### 2.2 计算性别与生还的相关系数

```python
df['Survived'].corr(df['Sex'], method = 'kendall')
```

生还与否和性别的相关性不强，相关系数仅供参考

#### 2.3 计算性别与生还的数据

```python
# 计算男女人数在Series里
df_count_sex = df.groupby('Sex')['PassengerId'].count()
df_count_sex
```

```python
# 计算男女性别各自的生还人数,数据放在Series里
sur_sex_count = df[df['Survived'] == 1].groupby('Sex')['PassengerId'].count()
sur_sex_count
```

```python
# 计算男女乘客的生还率，数据放在Series里
sur_rate_sex = sur_sex_count / df_count_sex
sur_rate_sex
```

#### 2.4 可视化性别和生还率

```python
ax1 = plt.subplot(121)
sur_sex_count.plot(kind = 'bar', color = 'green')
ax1.set_ylabel('生还人数')

ax2 = plt.subplot(122)
sur_rate_sex.plot(kind = 'bar', color = 'blue')
ax2.set_ylabel('生还率')

plt.subplots_adjust(left = 0.2, right = 1, wspace = 0.8)
```

从图中可以看出，女性的生还人数比男性多，女性的生还率也比男性高

>结论:
>1. 船上的男性乘客要多于女性乘客
>2. 生还与否于性别的相关系数较小
>3. 女性乘客获救的数量比男性多
>4. 女性的生还率比男性高

# 问题3：不同的票等级(pclass)与乘客生还的情况？

#### 3.1 三个票等级的乘客的分布情况

```python
count_each_pclass = pd.Series(df.groupby(['Pclass'])['PassengerId'].count().values, 
                        index = ['1-Upper','2-Middle','3-Lower'])
count_each_pclass
```

```python
count_each_pclass.plot(kind = 'pie', autopct='%.1f%%', explode = [0.05, 0.05, 0.05], shadow = True)

plt.title('三个票等级的乘客分布情况')
```

可以看出，所有乘客中，票等级3(Lower)的乘客比例最高，超过一半

3.2 计算生还与票等级的相关系数

```python
df['Survived'].corr(df['Pclass'], method = 'kendall')
```

生还与票等级相关系数小，相关系数仅供参考

#### 3.3 计算三种票等级的乘客的数据

```python
# 计算三种票等级的生还人数
sur_pclass_count =  pd.Series(df[df['Survived'] == 1].groupby('Pclass')['PassengerId'].count().values,
                             index = ['1-Upper','2-Middle','3-Lower'] )
```

```python
# 计算三种票等级的生还率
sur_rate_pclass = sur_pclass_count / count_each_pclass
sur_rate_pclass
```

#### 3.4 可视化票等级和生还率

```python
# 柱状图
ax1 = plt.subplot(121)
sur_pclass_count.plot(kind = 'bar', xlabel = '票等级', ylabel = '生还人数')

ax2 = plt.subplot(122)
sur_rate_pclass.plot(kind = 'bar', xlabel = '票等级', ylabel = '生还率')

plt.subplots_adjust(right = 1.5)
```

从条形图看，票等级为1的乘客生还人数最多，生还率也最高

>结论：
>1. 票等级为3的乘客占比最高
>2. 生还与票等级相关系数小
>3. 等级为1的乘客生还人数最多
>4. 等级为1的乘客生还率最高

# 问题4：不同登陆港口的乘客与生还情况如何？

#### 4.1 几个港口登录的乘客分布情况

```python
df_total_count_em = df.groupby('Embarked')['PassengerId'].count()
df_total_count_em
```

```python
df_total_count_em.plot(kind = 'pie', autopct = '%.1f%%', explode = [0.05, 0.05, 0.05], ylabel = '', shadow = True)
plt.axis('equal')
plt.show()
```

由上图可知：从S港口登录的乘客最多，占到了70%以上

#### 4.2 计算港口与生还的相关系数

```python
df['Embarked'].corr(df['Survived'], method = 'kendall')
```

生还与港口的相关系数小，相关度很弱，相关系数仅供参考

#### 4.3 计算各个港口的生还数据

```python
sur_count_em = df[df['Survived'] == 1].groupby('Embarked')['PassengerId'].count()
sur_count_em
```

#### 4.4 计算各个港口的生还率

```python
sur_rate_em = sur_count_em / df_total_count_em
sur_rate_em
```

#### 4.5 可视化各个港口登录的乘客生还数据

```python
ax1 = plt.subplot(121)
sur_count_em.plot(kind = 'bar', xlabel = '港口', ylabel = '生还数')

ax2 = plt.subplot(122)
sur_rate_em.plot(kind = 'bar', xlabel = '港口', ylabel = '生还率')

plt.subplots_adjust(left = 0.5, right = 1.5)
```

从上图可以看出，S港口的生还人数最多，C港口的生还率最高

>结论:
>1. 从S港口登录的人数最多
>2. 港口和生还的相关系数不高
>3. S港口的生化人数最多
>4. C港口的生还率最高

# 问题5：费用与票价等级的对应关系如何？ 

#### 5.1 费用

```python
# 费用的直方图
df['Fare'].hist()

# 设置标题
plt.title('Hist plot of Fare')

# 消除网格线
plt.grid(False)
```

大部分乘客的费用很低，在0-100之间

#### 5.2 可视化：票等级与费用的箱线图

```python
# 画箱线图
fare_pclass_1 = df[df['Pclass'] == 1]['Fare']
fare_pclass_2 = df[df['Pclass'] == 2]['Fare']
fare_pclass_3 = df[df['Pclass'] == 3]['Fare']
plt.boxplot([fare_pclass_1, fare_pclass_2, fare_pclass_3], labels=['1-Upper', '2-Middle', '3-Lower'])
plt.ylabel('Fare', fontsize = 12)
plt.grid(True)
plt.show()
```

从上图可以看出，票等级不同的乘客的费用是不同的，总体上1级的平均费用最高，其次是2级，3级最低

# 问题6：不同费用的乘客的生还情况如何？

```python
# 计算生还和死亡的乘客的费用
fare_sur = df[df['Survived'] == 1]['Fare']
fare_death = df[df['Survived'] == 0]['Fare']
fare_sur, fare_death
```

```python
# 画箱线图
plt.boxplot([fare_sur, fare_death], labels = ['生还乘客', '死亡乘客'])

# 设置y轴边界值
plt.ylim([-20, 550])
plt.ylabel('费用', fontsize = 20)
plt.grid(True)
plt.show()
```

从箱线图看，生还的乘客总体上费用较高

# 问题7：从三个港口登陆的乘客的费用是否有不同？

```python
# 不同港口的乘客的费用
fare_port_c = df[df['Embarked'] == 'C']['Fare']
fare_port_q = df[df['Embarked'] == 'Q']['Fare']
fare_port_s = df[df['Embarked'] == 'S']['Fare']
```

```python
plt.boxplot([fare_port_c, fare_port_q, fare_port_s], labels = ['C', 'Q', 'S'])
plt.xlabel('港口名')
plt.ylabel('费用')
plt.title('费用 VS 港口')
plt.grid(True)
```

总体上，从C港口登录的乘客花费较高
